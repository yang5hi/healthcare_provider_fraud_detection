```python
import pandas as pd


# load the data from csv to pandas dataframe
Healthcare_Inpatient_raw  = pd.read_csv('/Users/cuicuisong/Desktop/data/inpatient.csv')
Healthcare_Outpatient_raw= pd.read_csv('/Users/cuicuisong/Desktop/data/outpatient.csv')
Healthcare_Beneficiary_raw = pd.read_csv('/Users/cuicuisong/Desktop/data/beneficiary.csv')
Healthcare_Provider_raw  = pd.read_csv('/Users/cuicuisong/Desktop/data/provider.csv')
```

#### I. Merge Data

```python
# union/concat/merge the inpatient and outpatient data
concat_df=pd.concat([Healthcare_Inpatient_raw,Healthcare_Outpatient_raw],axis=0)
concat_df.head()
```

```python
# merge with beneficiary data
merge_bene_df=concat_df.merge(Healthcare_Beneficiary_raw, on='BeneID', how='left')
merge_bene_df.head()
```

```python
# merge with provider data
merge_provider_df=merge_bene_df.merge(Healthcare_Provider_raw, on = 'Provider', how ='left')
merge_provider_df.drop_duplicates(inplace=True)
merge_provider_df
```

#### II. Group by Provider: State, county, IPAnnualReimbursementAmt and IPAnnualDeductibleAmt

```python
# count of unique state, major state
# create a state-county feature and count of unique values (*double check the value).
# We can join and see if there are duplicate counties betwen states.
# IPAnnualReimbursementAmt and IPAnnualDeductibleAmt: Mean, Max, Sum
```

###### 1.1. State_unique

```python
statedf= pd.DataFrame(merge_provider_df.groupby(['Provider']).State.nunique())
statedf=statedf.rename(columns={"State": "statecount_uniq"})
statedf
```

```python
provider_scc_df= Healthcare_Provider_raw
```

```python
# merge statecount_uniq with provider data.
provider_scc_df=provider_scc_df.merge(statedf, on = 'Provider', how ='left')
provider_scc_df
```

##### 1.2 State_major

```python
#statedf_max= pd.DataFrame(merge_provider_df.groupby(['Provider']).State.max())
#statedf_max=statedf_max.rename(columns={"State": "state_max"})
```

```python
# count the frequent of state and keep the first row (the most frequent state)
statedf_major= pd.DataFrame(merge_provider_df.groupby(['Provider']).State.apply(lambda x: x.value_counts().head(1))\
               .reset_index())
statedf_major
```

```python
# drop state count column and rename the level_1 to major_state
statedf_major=statedf_major[['Provider','level_1']]
statedf_major=statedf_major.rename(columns={"level_1": "major_state"})
statedf_major
```

```python
# add column state_max to provider data.
provider_scc_df=provider_scc_df.merge(statedf_major, on = 'Provider', how ='left')
provider_scc_df
```

##### 2.1 County_unique

```python
# combine state and county
merge_provider_df['countyfips']=merge_provider_df['State'].astype(str)+Healthcare_Beneficiary_raw['County'].astype(str)
```

```python
countydf= pd.DataFrame(merge_provider_df.groupby(['Provider']).County.nunique())
countydf=countydf.rename(columns={"County": "county_uniq"})
countydf
```

```python
# add county_uniq to provider data.
provider_scc_df=provider_scc_df.merge(countydf, on = 'Provider', how ='left')
provider_scc_df
```

##### 2.2 County_major

```python
countydf_major= pd.DataFrame(merge_provider_df.groupby(['Provider']).County.apply(lambda x: x.value_counts().head(1))\
               .reset_index())
countydf_major
```

```python
# drop county count column and rename the level_1 to major_county
countydf_major=countydf_major[['Provider','level_1']]
countydf_major=countydf_major.rename(columns={"level_1": "major_county"})
countydf_major
```

```python
# add county_major to provider data.
provider_scc_df=provider_scc_df.merge(countydf_major, on = 'Provider', how ='left')
provider_scc_df
```

#### 3. IPAnnualReimbursementAmt
##### 3.1 sum_IPAnnualReimbursementAmt

```python
IPAnnualReimbursementAmtdf = pd.DataFrame(merge_provider_df.groupby(['Provider']).IPAnnualReimbursementAmt.sum())  
IPAnnualReimbursementAmtdf=IPAnnualReimbursementAmtdf.rename(columns={"IPAnnualReimbursementAmt": "sum_IPAnnualReimbursementAmt"})
IPAnnualReimbursementAmtdf
```

```python
provider_scc_df=provider_scc_df.merge(IPAnnualReimbursementAmtdf, on = 'Provider', how ='left')
provider_scc_df
```

##### 3.2 mean_IPAnnualReimbursementAmt

```python
IPAnnualReimbursementAmtdf_mean = pd.DataFrame(merge_provider_df.groupby(['Provider']).IPAnnualReimbursementAmt.mean())  
IPAnnualReimbursementAmtdf_mean = IPAnnualReimbursementAmtdf_mean.rename(columns={"IPAnnualReimbursementAmt": "mean_IPAnnualReimbursementAmt"})
IPAnnualReimbursementAmtdf_mean
```

```python
provider_scc_df=provider_scc_df.merge(IPAnnualReimbursementAmtdf_mean, on = 'Provider', how ='left')
provider_scc_df
```

#####  3.3 max_IPAnnualReimbursementAmt

```python
IPAnnualReimbursementAmtdf_max = pd.DataFrame(merge_provider_df.groupby(['Provider']).IPAnnualReimbursementAmt.max())  
IPAnnualReimbursementAmtdf_max = IPAnnualReimbursementAmtdf_max.rename(columns={"IPAnnualReimbursementAmt": "max_IPAnnualReimbursementAmt"})
IPAnnualReimbursementAmtdf_max
```

```python
provider_scc_df=provider_scc_df.merge(IPAnnualReimbursementAmtdf_max, on = 'Provider', how ='left')
provider_scc_df
```

#### 4. IPAnnualDeductibleAmt:Sum, Mean, Max
##### 4.1 Sum_IPAnnualDeductibleAmt

```python
IPAnnualDeductibleAmtdf = pd.DataFrame(merge_provider_df.groupby(['Provider']).IPAnnualDeductibleAmt.sum())  
IPAnnualDeductibleAmtdf = IPAnnualDeductibleAmtdf.rename(columns={"IPAnnualDeductibleAmt": "sum_IPAnnualDeductibleAmtdf"})
IPAnnualDeductibleAmtdf
```

```python
provider_scc_df= provider_scc_df.merge(IPAnnualDeductibleAmtdf, on = 'Provider', how ='left')
provider_scc_df
```

##### 4.2 mean_IPAnnualDeductibleAmt

```python
IPAnnualDeductibleAmtdf_mean = pd.DataFrame(merge_provider_df.groupby(['Provider']).IPAnnualDeductibleAmt.mean())  
IPAnnualDeductibleAmtdf_mean=IPAnnualDeductibleAmtdf_mean.rename(columns={"IPAnnualDeductibleAmt": "mean_IPAnnualDeductibleAmt"})
IPAnnualDeductibleAmtdf_mean
```

```python
provider_scc_df= provider_scc_df.merge(IPAnnualDeductibleAmtdf_mean, on = 'Provider', how ='left')
provider_scc_df
```

##### 4.3 max_IPAnnualDeductibleAmt

```python
IPAnnualDeductibleAmtdf_max = pd.DataFrame(merge_provider_df.groupby(['Provider']).IPAnnualDeductibleAmt.max())  
IPAnnualDeductibleAmtdf_max=IPAnnualDeductibleAmtdf_max.rename(columns={"IPAnnualDeductibleAmt": "max_IPAnnualDeductibleAmt"})
IPAnnualDeductibleAmtdf_max
```

```python
provider_scc_df= provider_scc_df.merge(IPAnnualDeductibleAmtdf_max, on = 'Provider', how ='left')
provider_scc_df
```
